```python



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df = pd.read_csv(url)
print(df.head())
```

```python
# -----------------------------
# DATASET DESCRIPTION
# -----------------------------
print("\nDataset Info:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
# -----------------------------
# DATA SCHEMA (COLUMNS)
# -----------------------------
print("\nColumns in Dataset:")
for col in df.columns:
    print(col)
```

```python
print("\nMean Values:")
print(df.mean(numeric_only=True))
```

```python
print("\nVariance Values:")
print(df.var(numeric_only=True))
```

```python
print("\nStandard Deviation:")
print(df.std(numeric_only=True))
```

```python


print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
print(" DATA CLEANING & QUALITY CHECKS ")
print("\nMissing Values per Column:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nData Quality Observations:")
if df.isnull().sum().sum() > 0:
    print(" -> Alert: Dataset contains missing values. Consider filling (fillna) or removing (dropna) them.")

if df.duplicated().sum() > 0:
    print(" -> Alert: Dataset contains duplicate rows. Consider removing them (drop_duplicates).")
```

```python
-

# Histogram for numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
```

```python
for col in numeric_cols:
    plt.figure()
    plt.hist(df[col].dropna())
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python

categorical_cols = df.select_dtypes(include='object').columns
```

```python
for col in categorical_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python

print("\nBasic Insights:")
for col in numeric_cols:
    print(f"Average {col}:", df[col].mean())

for col in categorical_cols:
    print(f"Most common {col}:", df[col].mode()[0])
```

```python

print("\nData Quality Observations:")

if df.isnull().sum().sum() > 0:
    print("Dataset contains missing values. Consider filling or removing them.")

if df.duplicated().sum() > 0:
    print("Dataset contains duplicate rows. Consider removing duplicates.")

print("Check for outliers using visual inspection (histograms).")
```
